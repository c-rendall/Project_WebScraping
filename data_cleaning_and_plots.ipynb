```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
%matplotlib inline 
```

```python
#Load csv file as dataframe
df = pd.read_csv('courses_whole3.csv')
df
```

```python
#Remove 'stars' from the rating column
df['rating'] = df['rating'].str.replace('\nstars', '')
df
```

```python
#Remove commas from the enrollment column
df['enrollment'] = df['enrollment'].str.replace(',', '')
df
```

```python
#Remove 'Missing' from rating and convert to NaN
df['rating'] = pd.to_numeric(df.rating, errors='coerce')
```

```python
df
```

```python
#Top 20 partners by number of courses. This only shows partners that were not missing.  Missing values accounted
#for 244 of the total, which is the largest count.  This can be shown by changing [1:21] to [:21]
top_20_partners = df.partner.value_counts()
top_20_partners[1:21].plot(kind='barh')
plt.xlabel('Number of Courses')
```

```python
#Top subjects by number of courses
top_subjects = df.subject.value_counts()
top_subjects[:11].plot(kind='barh')
plt.xlabel('Number of Courses')
```

```python
#Average rating by subject
rating_by_subject = df.groupby('subject')['rating'].mean().sort_values(ascending=False)
rating_by_subject
```

```python
#subject_avg_rating = df.groupby('subject')['rating'].boxplot()
fig = plt.figure(figsize=(10,8))
ax = fig.gca()
df.boxplot(column='rating', by='subject', ax=ax, vert=False)
ax.set_title('Boxplots of Rating by Subject')
ax.set_xlabel('Rating')
ax.set_ylabel('Subject')
```

```python
#Top 10 worst rated courses 
top10_worst_classes = df.groupby('title')['rating'].mean().sort_values()
top10_worst_classes[:20].plot(kind='barh')
plt.xlabel('Rating')
```

```python
worst_classes_subject_rating = df.groupby('subject')['rating', 'enrollment'].mean().sort_values(by='enrollment')
worst_classes_subject_rating
```

```python
#Dataframe filtered for only the skills learned in data science courses
skills = df['skills'].str[:]
subject = df['subject']
subject_skills = pd.concat([subject, skills], axis=1)
data_science = subject_skills.loc[subject_skills['subject'] == 'Data Science']
data_science
```

```python
#Rating column boxplot
rating_boxplot = sns.boxplot(x='rating', 
                 data=df, 
                 width=0.5,
                 palette="colorblind")
```

```python
#Count the number of occurences of a class having a rating below 4.0.  We can check above a 4.0 by flipping the 
#less-than symbol
def sum_ratings(df):
    return (df.rating.values < 4.0).sum()

sum_ratings(df)
```

```python
# Pie chart of low-rated classes
labels = 'Below 4.0', 'Above 4.0'
sizes = [69, 2460]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
