```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
import time
import csv
import re
import math

#Open a blank csv
csv_file = open('courses_whole.csv', 'w')

writer = csv.writer(csv_file)

#Create column headers for the variables
writer.writerow(['title', 'partner', 'rating', 'enrollment', 
                 'subject', 'skills', 'duration', 'level', 
                 'new_career', 'promotion'])
```

```python
driver = webdriver.Chrome()

#Main Coursera page filtered for courses conducted in English only
driver.get('https://www.coursera.org/search?index=prod_all_products_term_optimization&allLanguages=English&entityTypeDescription=Courses')

#Scrape the total number of courses after being filtered
num_classes_str = driver.find_element_by_xpath('//h3[@class="rc-NumberOfResultsSection body-2-text"]').text

#Convert number of classes to an integer
num_classes = int((re.findall(r'\d+', num_classes_str))[0])

#Initialize page number to 1
current_page = 1  

#Calculate total number of pages
num_pages = math.ceil(num_classes / 10)

#Course urls
course_urls = []

# Initiate while loop to scroll through all pages
while current_page <= num_pages:
    time.sleep(2)
    a_tags = driver.find_elements_by_xpath('//li[@class="ais-InfiniteHits-item"]/div/a')
    hrefs = [a_tag.get_attribute('href') for a_tag in a_tags]
    course_urls.extend(hrefs)
    current_page += 1
    driver.execute_script("window.scrollTo(0, 1500);")
    button = driver.find_element_by_xpath('//button[@aria-label="Next Page"]')
    button.click()
    
```

```python
#Check that all urls were saved
len(course_urls)
```

```python
#Title of course
def get_title():
    try:
        title = driver.find_element_by_xpath('//h1[@class="banner-title m-b-0"]')
    except:
        title = 'Missing'
    finally:
        return title 

#Partner sponsoring the course
def get_partner():
    try:
        title = driver.find_element_by_xpath('//div[@class = "m-b-1s m-r-1img"]//img.get_atttribute("title")')
    except:
        title = 'Missing'
    finally:
        return title

#Course rating on scale of 0-5    
def get_rating():
    try:
        rating = driver.find_element_by_xpath('//span[@class="_16ni8zai m-b-0 rating-text number-rating number-rating-expertise"]')
    except:
        rating = 'Missing'
    finally:
        return rating

#Total number of people who previously enrolled
def get_enrollment():
    try:
        enrollment = driver.FindElement(By.XPath('//div[@class="_3vb6hs"]/span/strong'))
    except:
        enrollment = 'Missing'
    finally:
        return enrollment

#Subject of the course
def get_subject():
    try:
        subject = driver.FindElement(By.XPath('//div[@class="_1ruggxy"]/a'))
    except:
        subject = 'Missing'
    finally:
        return subject

#Skills acquired by taking the course
def get_skills():
    try:
        skills = driver.FindElement(By.Xpath('//div[@class="Skills p-x-2 p-t-1 p-b-2 skills-sdp-content-exp border-a"][@class="content-inner"]'))
    except:
        skills = 'Missing'
    finally:
        return skills

#Duration of the course
def get_duration():
    try:
        duration = driver.find_element_by_xpath('//div[@class="_1tu07i3a"]/h4[@class="_16ni8zai m-b-0"]/span')
    except:
        duration = 'Missing'
    finally:
        return duration

#Course difficulty level: beginner, intermediate, or advanced
def get_level():
    try:
        level = driver.find_element_by_xpath('//div[@class="_1tu07i3a"]/h4[@class="_16ni8zai m-b-0"]')
    except:
        level = 'Missing'
    finally:
        return level

#Percent of people who started a new career after taking the course
def get_newcareer():
    try:
        new_career = driver.find_element_by_xpath('//div[@class="_1k3yl1y"]//h2[@class="_1l2q8kho m-x-1 m-b-0"]/h2')
    except:
        new_career = 'Missing'
    finally:
        return new_career

#Percent of people who got a raise or promotion after 
def get_promotion():
    try:
        promotion = driver.find_element_by_xpath('//div[@class="_jyhj5r"]//h2[@class ="_1l2q8kho m-x-1 m-b-0"]/h2')
    except:
        promotion = 'Missing'
    finally:
        return promotion
    

    course_dict = {}
    course_dict['title'] = get_title()
    course_dict['partner'] = get_partner()
    course_dict['rating'] = get_rating()
    course_dict['enrollment'] = get_enrollment()
    course_dict['subject'] = get_subject()
    course_dict['skills'] = get_skills()
    course_dict['duration'] = get_duration()
    course_dict['level'] = get_level()
    course_dict['new_career'] = get_newcareer()
    course_dict['promotion'] = get_promotion()
    writer.writerow(course_dict.values())
    driver.close()
    
driver = webdriver.Chrome()
for url in course_urls:
    course_dict = {}
    driver.get(url)
    
#Close the csv file
csv_file.close()
```
